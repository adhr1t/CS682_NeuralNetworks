The kNN classifier consists of two stages:

- During training, the classifier takes the training data and simply remembers it
- During testing, kNN classifies every test image by comparing to all training images and transfering the labels of the k most similar training examples
- The value of k is cross-validated

We would now like to classify the test data with the kNN classifier. Recall that we can break down this process into two steps:

1. First we must compute the distances between all test examples and all train examples.

**Your Answer**: *fill this in.*

- When there are high distances between the train image and test image, the row is bright white. This means there are very little pixel-based/numerical similarities between the images, and the calculated Euclidean distance is very high. One row is a single test example, so when a row is very bright white, it means the single test example has a very high Euclidean distance from a large amount of the training examples. This could mean the test example is pixel-wise/numerically very different image from all the training examples and the model might have a difficult time calculating a nearest neighbor. Or the test example might not belong to a class in the training examples.

- Because each row is a single test example, every column is one of the many training examples. Thus, the grid is a result of the test examples intersecting with every single training example. Bright white columns occurs if a particular training example is pixel-wise/numerically very different image from all of the testing examples and the model might have a difficult time calculating a nearest neighbor.

```python
# Now lets speed up distance matrix computation by using partial vectorization
# with one loop. Implement the function compute_distances_one_loop and run the
# code below:
dists_one = classifier.compute_distances_one_loop(X_test)

# To ensure that our vectorized implementation is correct, we make sure that it
# agrees with the naive implementation. There are many ways to decide whether
# two matrices are similar; one of the simplest is the Frobenius norm. In case
# you haven't seen it before, the Frobenius norm of two matrices is the square
# root of the squared sum of differences of all elements; in other words, reshape
# the matrices into vectors and compute the Euclidean distance between them.
difference = np.linalg.norm(dists - dists_one, ord='fro')
print('Difference was: %f' % (difference, ))
if difference < 0.001:
    print('Good! The distance matrices are the same')
else:
    print('Uh-oh! The distance matrices are different')
```

```python
# Let's compare how fast the implementations are
def time_function(f, *args):
    """
    Call a function f with args and return the time (in seconds) that it took to execute.
    """
    import time
    tic = time.time()
    f(*args)
    toc = time.time()
    return toc - tic

two_loop_time = time_function(classifier.compute_distances_two_loops, X_test)
print('Two loop version took %f seconds' % two_loop_time)

one_loop_time = time_function(classifier.compute_distances_one_loop, X_test)
print('One loop version took %f seconds' % one_loop_time)

no_loop_time = time_function(classifier.compute_distances_no_loops, X_test)
print('No loop version took %f seconds' % no_loop_time)

# you should see significantly faster performance with the fully vectorized implementation
```

```python
num_folds = 5
k_choices = [1, 3, 5, 8, 10, 12, 15, 20, 50, 100]

X_train_folds = []
y_train_folds = []
################################################################################
# TODO:                                                                        #
# Split up the training data into folds. After splitting, X_train_folds and    #
# y_train_folds should each be lists of length num_folds, where                #
# y_train_folds[i] is the label vector for the points in X_train_folds[i].     #
# Hint: Look up the numpy array_split function.                                #
################################################################################
X_train_folds = np.array_split(X_train, num_folds) # length of 5 (num_folds)
y_train_folds = np.array_split(y_train, num_folds) # length of 5 (num_folds)
################################################################################
#                                 END OF YOUR CODE                             #
################################################################################

# A dictionary holding the accuracies for different values of k that we find
# when running cross-validation. After running cross-validation,
# k_to_accuracies[k] should be a list of length num_folds giving the different
# accuracy values that we found when using that value of k.
k_to_accuracies = {}
# k is the # of nearest neighbors of training images we want the test point to be near

################################################################################
# TODO:                                                                        #
# Perform k-fold cross validation to find the best value of k. For each        #
# possible value of k, run the k-nearest-neighbor algorithm num_folds times,   #
# where in each case you use all but one of the folds as training data and the #
# last fold as a validation set. Store the accuracies for all fold and all     #
# values of k in the k_to_accuracies dictionary.                               #
################################################################################

for k in k_choices:



  #   # store k as key and accuracy as value in dictionary
  # k_to_accuracies[k] = accuracy

  # run knn 5 times (or whatever num_folds is)
  # each run uses 4 training folds and 1 changing Validation fold
  # pop the first fold to the end, train on 4 folds, validate on the last fold.
  # Repeat 5 times


  # this list stores the accuracies for each orientation of the folds.
  # It then resets when all orientations have been exhausted
  accuracy = []

  # ## swapping the folds for training/validation
  # # need to np.roll() num_folds - 1 times (in this case 4 times)
  for n in range(num_folds):
  #   # print('n', n)


    ## recombining X_train_fold_four
    # rolls first array element/fold to the end
    # X_train_folds_five = np.roll(X_train_folds, 1)
    # y_train_folds_five = np.roll(y_train_folds, 1)

    # stores the first 4 (num_folds - 1) elements/folds in X_train_fold_four/y_train_fold_four
    # Need to concatenate the folds. The extra dimension is throwing it off
    # X_train_fold_four = X_train_folds_five[:num_folds-1].reshape(-1, 3072)
    # print("X_train_fold_four", X_train_fold_four)
    # # print(X_train_fold_four.shape)
    # # print(X_train_fold_four.transpose(2,0,1).reshape(-1,X_train_fold_four.shape[1]).shape)
    # # print(X_train_fold_four.transpose(2,0,1).reshape(-1, 3072).shape)
    # y_train_fold_four = y_train_folds_five[:num_folds-1]

    # # captures the last element of X_train_folds/y_train_folds array for validation
    # X_val_fold = X_train_folds_five[num_folds - 1]
    # y_val_fold = y_train_folds_five[num_folds - 1]


    ## what if i shift my validation set with n ??
    # capture the nth fold with the loop of n
    x_val_fold = X_train_folds[n]
    y_val_fold = y_train_folds[n]

    # capture every other fold as the training folds
    # print(X_train_folds[:0])  # runs fine if n is 0
    # print(X_train_folds[4:])  # runs fine if n is 4. Both of these
    # examples are at the edges of the array and run fine, so I don't need
    # the if statement to watch out for edge cases
    x_train_fold = np.concatenate(X_train_folds[:n] + X_train_folds[n + 1:])
    # print("x_train_fold", x_train_fold)
    y_train_fold = np.concatenate(y_train_folds[:n] + y_train_folds[n + 1:])



    # instantiate KNN
    knn = KNearestNeighbor()
    # feed in 4 training folds
    knn.train(x_train_fold, y_train_fold)
    # feed in 1 validation fold and save outputted labels to y_val_pred
    y_val_pred = knn.predict(X = x_val_fold, k = k, num_loops = 0)

    # calculate KNN's accuracy for this orientation of the folds
    num_correct = np.sum(y_val_pred == y_val_fold)
    acc = float(num_correct) / num_test

    # append the current fold orientation's validation accuracy to the accuracy
    accuracy.append(acc)

    # store k as key and accuracy as value in dictionary
  k_to_accuracies[k] = accuracy



################################################################################
#                                 END OF YOUR CODE                             #
################################################################################

# Print out the computed accuracies
for k in sorted(k_to_accuracies):
    for accuracy in k_to_accuracies[k]:
        print('k = %d, accuracy = %f' % (k, accuracy))
```

```python
# plot the raw observations
for k in k_choices:
    accuracies = k_to_accuracies[k] # looks like k_to_accuracies[k] is associated with multiple accuracies of diff fold orders
    plt.scatter([k] * len(accuracies), accuracies)

# plot the trend line with error bars that correspond to standard deviation
accuracies_mean = np.array([np.mean(v) for k,v in sorted(k_to_accuracies.items())])
accuracies_std = np.array([np.std(v) for k,v in sorted(k_to_accuracies.items())])
plt.errorbar(k_choices, accuracies_mean, yerr=accuracies_std)
plt.title('Cross-validation on k')
plt.xlabel('k')
plt.ylabel('Cross-validation accuracy')
plt.show()
```
